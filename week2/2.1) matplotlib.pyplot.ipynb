# matplotlib.pyplot 

+ for creaing a figure or an area, plotting some lines(graphs) or decorating the plot with labels, etc. 

```python
from matplotlib import pyplot as plt
import numpy as np
```

### figure() & show()
+ ***plt.figure()*** creates new figure.
+ ***plt.show()*** shows all figure.

```python
# plt.show() shows nothing.
# Becuase, we've just created the "figure" instance.

plt.figure()
plt.show()
```

### figure size
+ we can set figure size.
+ below example sets the figure size 9inch*9inch

```python
plt.figure(figsize=(9, 9))
```

### axis()
+ plt axis() works with figure() function.
+ In the example below, axis() function takes a list of **[xmin, xmax, ymin, ymax]** and specifies the viewport of the axes.

```python
plt.axis([0, 6, 0, 20])
plt.show()
```

### plot()
+ plot() functions helps us to create graph. 
</br>
+ If you provide a single list or array to plot, matplotlib assumes it is a sequence of y values, and automatically creates it

```python
x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])

# automatically creates the sequence of y value.
plt.plot(x)
plt.show()
```

```python
plt.plot(x, y)
plt.show()
```

### optional styles

```python
plt.plot(x, y, color='red', linewidth=4)
plt.show()
```

```python
plt.plot(x, y, 'ro')
plt.show()
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### Scatter()

```python
plt.figure()
 
x = np.array([0.3, 3.8, 1.2, 2.5])
y = np.array([11, 25, 9, 26])
 
plt.scatter(x, y, marker='^', color='darkgreen')
plt.show()
```

### add strings
+ suptitle()
+ xlabel(), ylabel()

```python
plt.axis([0, 6, 0, 20])
plt.title('show the title', fontsize=20)


plt.xlabel('$x_1$')
plt.ylabel('$y_1$', fontsize = 15)
plt.show()
```

### Subplot()
+ By using Subplot(), you can draw other graphs in one figure.
+ subplot(131) means in dived the figure in **1*3**, and choose **1**(the first one).
+ subplot(132) means in dived the figure in **1*3**, and choose **2**(the second one).
+ subplot(133) means in dived the figure in **1*3**, and choose **3**(the third one).

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 4))

plt.subplot(131)
plt.bar(names, values)
plt.title('bar')

plt.subplot(132)
plt.scatter(names, values)
plt.title('scatter')

plt.subplot(133)
plt.plot(names, values)
plt.title('plot')

plt.suptitle('Categorical Plotting', fontsize=15)
plt.show()
```
